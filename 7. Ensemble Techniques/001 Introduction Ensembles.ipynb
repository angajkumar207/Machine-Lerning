### What is an ensemble?
- An ensemble combines multiple models (called base learners or weak learners) so the combined prediction is better (more accurate, robust, or calibrated) than any  single model. The key is diversity among base learners and a good combination method.

- 

### **Main ensemble families**

 1. Bagging (Bootstrap Aggregating) — train many models on random bootstrap samples; combine by averaging (regression) or majority vote (classification). Example: Random Forest.
 2. Boosting — sequentially train models where each new model focuses on previous errors. Models are combined with weighted sums. Example: AdaBoost, Gradient Boosting, - XGBoost, LightGBM.
 3. Stacking (Stacked Generalization) — train different base models, then train a meta-model on their predictions to learn the best combination.
